# loading the sample video

```python

import torch
import torchvision
from torchvision.models.detection import FasterRCNN 
from torchvision.transforms import functional as F 
import cv2
# Load a pre-trained Faster R-CNN model
model = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained=True) 
model.eval()
# Load an image
image = cv2.imread('download.jpeg')
# Preprocess the image
image_tensor = F.to_tensor(image).unsqueeze (0)
# Perform object detection
with torch.no_grad():
    predictions = model(image_tensor)
    
boxes = predictions[0]['boxes']
scores = predictions[0]['scores']
labels = predictions[0]['labels']

# Process detection results boxes = predictions[®]['boxes'] scores = predictions[@]['scores'] labels = predictions[0]['labels']
# Draw bounding boxes on the image
for box, score, label in zip(boxes, scores, labels):
        box = box.tolist()
        x1, y1, x2, y2 = map(int, box)
        cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(image, f'Class: {label}, Score: {score:.2f}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
        # Display the result
        cv2.imshow('Faster R-CNN Object Detection', image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
```
